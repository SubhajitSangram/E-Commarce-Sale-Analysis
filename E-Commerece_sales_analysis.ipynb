```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import plotly.colors as colors
pio.templates.default = "plotly_white"
```

```python
data=pd.read_csv('Sample_Superstore.csv',encoding='latin-1')
```

```python
data[['Order Date']].head()
```

```python
#Split Column and new cloemn add(Day, Month, Year)
data['Order_Day']=data['Order Date'].str.split('/').str[0]
data['Order_Month']=data['Order Date'].str.split('/').str[1]
data['Order_Year']=data['Order Date'].str.split('/').str[2]
```

```python
data[['Order Date','Order_Day','Order_Month','Order_Year']].head()
```

```python
#Change Datatype
data['Order_Day']=data['Order_Day'].astype(int)
data['Order_Month']=data['Order_Month'].astype(int)
data['Order_Year']=data['Order_Year'].astype(int)
```

```python
data[['Order Date','Order_Day','Order_Month','Order_Year']].info()
```

```python
#1)You need to calculate the monthly sales of the store and identify which month had the highest sales and which 
# month had the lowest sales.
sales_by_month=data.groupby('Order_Month')['Sales'].sum().reset_index()
```

```python
# Plot manually
Plot=px.line(
    sales_by_month,
    x='Order_Month',
    y='Sales',
    title='Month By Sales'
)
Plot.show()
```

#### Highest Sales Month: 17.0 - 112.1551K
#### Lowest Sales Month: 29.0 - 44.80669K

```python
#2. You need to analyze sales based on product categories and determine which category has the
# lowest sales and which category has the highest sales.

sales_by_category=data.groupby('Category')['Sales'].sum().reset_index()
```

```python
data['Category'].unique()
```

```python
chart=px.pie(
    sales_by_category,
    names='Category',
    values='Sales',
    hole=0.5,
    title='Catagory By Sales'
)
chart.show()
```

#### Catagory has Highest Sales : Office Supplies (36.4%)
#### Catagory has Lowest Sales : Technology (31.3%)

```python
# 3. The sales analysis needs to be done based on sub-categories
sales_by_subcategory = data.groupby('Sub-Category')['Sales'].sum().reset_index()

bar = px.bar(sales_by_subcategory, 
             x='Sub-Category', 
             y='Sales',
             title='Sales Analysis by Sub-Category')
bar.show()
```

#### First Highest Sales Sub-Catagory : Phones (330.007k) and second highest sales :Chairs (328.4491k)
#### Last Highest Sales Sub-Catagory :Fasteners (3024.28)
 

```python
# 4. You need to analyze the monthly profit from sales and determine which month had the highest profit.

profit_by_month = data.groupby('Order_Month')['Profit'].sum().reset_index()
fig = px.line(profit_by_month, 
              x='Order_Month', 
              y='Profit', 
              title='Monthly Profit Analysis',
            
              )
fig.show()
```

#### Highest Month Profit : month=23, profit=20.0966k

```python
#5. Analyze the profit by category 
profit_by_category = data.groupby('Category')['Profit'].sum().reset_index()

fig = px.pie(profit_by_category, 
             values='Profit', 
             names='Category', 
             hole=0.5, 
             title='Category by Profit',          
             )
fig.show()
```

```python
#5. Analyze the profit by subcategory 
sales_by_subcategory = data.groupby('Sub-Category')['Sales'].sum().reset_index()
fig = px.bar(sales_by_subcategory, 
             x='Sub-Category', 
             y='Sales', 
             title='Sales Analysis by Sub-Category'
             )
fig.show()
```

```python
# 6. Analyze the sales and profit by customer segment
sales_profit_by_segment = data.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

fig = px.bar(
    sales_profit_by_segment,
    x='Segment',
    y=['Sales', 'Profit'],
    barmode='group',
    title='Sales and Profit by Customer Segment',
    labels={'value': 'Amount', 'Segment': 'Customer Segment'},
    text_auto=True,
    color_discrete_sequence=["#0EEEA7","#C37C54"]
)
fig.show()
```

```python
# 7. Analyze the sales to profit ratio
sales_profit_by_segment = data.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
sales_profit_by_segment['Sales_to_Profit_Ratio'] = sales_profit_by_segment['Sales'] / sales_profit_by_segment['Profit']
print(sales_profit_by_segment[['Segment', 'Sales_to_Profit_Ratio']])
```
